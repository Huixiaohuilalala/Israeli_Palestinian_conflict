```python
from gensim import corpora
from gensim.models import LdaModel
from gensim import models
from gensim.corpora import Dictionary
import pandas as pd
import matplotlib.pyplot as plt
from gensim.models import CoherenceModel
from pyLDAvis import gensim
import pyLDAvis
import numpy as np
```

```python
def map2list(x):
    return x.split()
```

```python
dataFrame = pd.read_csv('./ENdata/reddit_4.csv')
dataFrame
```

```python
texts = dataFrame['text'].map(map2list).values
dictionary = Dictionary(texts)
```

```python
# 压缩词向量，去掉出现的文章小于2的词，和在50%的文章都出现的词，整体长度限制在1000
dictionary.filter_extremes(no_below=2, no_above=0.5, keep_n=1000)
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
# 进行TF/IDF编码
tfidf = models.TfidfModel(corpus)
corpus_tfidf = tfidf[corpus]
```

```python
dictionary.token2id
```

```python
list(corpus_tfidf)[:10]
```

```python
# 训练LDA
num_topics = 14
print(f'开始训练第{num_topics}个主题')
# 训练LDA模型,passes为迭代次数
lda = LdaModel(corpus=corpus_tfidf, id2word=dictionary, num_topics=num_topics, passes=50)
print(f'训练第{num_topics}个主题完毕')
lda.save(f'./ENmodel/batch4/{num_topics}.model')
print(f'第{num_topics}个主题保存完毕')
# 打印主题
for topic in lda.print_topics(num_words=25, num_topics=num_topics):
    print(topic)
```
